```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("test01.txt", dtype=float).T
rho = data[0]
nH = data[1]
T = data[2]
etaO = data[-3]
etaH = data[-2]
etaA = data[-1]

etaHp = np.where(etaH >= 0, etaH, np.nan)
etaHm = np.where(etaH < 0, etaH, np.nan)

del data
```

```python
x = rho
plt.figure(figsize=(12, 10))
plt.plot(x, etaO)
plt.plot(x, etaA)
plt.plot(x, etaHp)
plt.plot(x, np.abs(etaHm))
plt.xscale("log")
plt.yscale("log")
plt.xlim(x[0], x[-1])
plt.ylim(1.0e17, 1.0e22)
plt.show()
```

```python
data = np.loadtxt("test01agg.txt", dtype=float).T
etaO2 = data[3]
etaH2 = data[4]
etaA2 = data[5]

etaH2p = np.where(etaH2 >= 0, etaH2, np.nan)
etaH2m = np.where(etaH2 < 0, etaH2, np.nan)

del data
```

```python
plt.figure(figsize=(12, 10))
plt.plot(x, etaO, label="etaO1", color="black")
plt.plot(x, etaA, label="etaA1", color="red")
plt.plot(x, etaHp, label="etaH1>0", color="blue")
plt.plot(x, np.abs(etaHm), label="etaH1<0", color="purple")
plt.plot(x, etaO2, label="etaO1", color="black", linestyle="dashed")
plt.plot(x, etaA2, label="etaA2", color="red", linestyle="dashed")
plt.plot(x, etaH2p, label="etaH2>0", color="blue", linestyle="dashed")
plt.plot(x, np.abs(etaH2m), label="etaH2>0", color="purple", linestyle="dashed")
plt.xscale("log")
plt.yscale("log")
plt.xlim(x[0], x[-1])
plt.ylim(1.0e17, 1.0e22)
plt.legend(fontsize=20)
plt.show()
```
